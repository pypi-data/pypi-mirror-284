```python
from burst_waveform.models import WhiteNoiseBurst

params = {
    'frequency': 2000,
    'bandwidth': 1000,
    'duration': 0.05,
    'inj_length': 1,
    'mode': 1
}

WNB = WhiteNoiseBurst(params)
wnb_strain = WNB()
```

```python
import matplotlib.pyplot as plt
plt.plot(wnb_strain)
plt.xlim(- 10 * params['duration'], 10 * params['duration'])
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
wnb_strain.to_pycbc().to_frequencyseries().plot()
plt.show()
```
